## 4. Summarizing Data

### a. Research Question 

- What is the Relationship between US Sales(miilions) and Genre of the game?

### b. Visualization 

```python
plt.figure(figsize = (14,6))
sns.boxplot(x = 'Genre', y = 'US Sales (millions)', data = df_cleaned)
plt.xticks(rotation = 90)
plt.title("Distribution of video games")
plt.xlabel('Genre')
plt.ylabel('US Sales (millions)')
plt.show()
```

### c. Summary statistics

```python
stats_genre = df_cleaned.groupby('Genre')['US Sales (millions)'].describe()
stats_genre
```

### d. Describe from numerical summary and visualization

- The visualization shows the distribution of video game sales between genres. 
From the boxplot, we can observe that some genres like Action, RPG, Sports, and Racing/Driving show the highest sales value, with several outliers exceeding 10 million. On the other hand, Strategy and Educational genres show the lower sales with fewer extreme values. When we look at the summary statistics, the median sales show that most genres have median value below 1 million, meaning that while a few games have enormous success, most remain in lower values. The **NaN** values that you notice are that the count of observation is just 1. So, the standard deviation would be zero, and min/max/IQR values would be same.


### e. Answer of the research question
